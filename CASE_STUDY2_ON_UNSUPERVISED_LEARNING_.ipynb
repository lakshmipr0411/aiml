```python
#Apply IQR method
cols=['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']
for c in cols:
    Q1=wholesale_customers[c].quantile(0.25)
    Q3=wholesale_customers[c].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    wholesale_customers[c]=np.where(wholesale_customers[c]<lower_bound,lower_bound,wholesale_customers[c])
    wholesale_customers[c]=np.where(wholesale_customers[c]>upper_bound,upper_bound,wholesale_customers[c])
```

K-Means Clustering

```python
from sklearn.cluster import KMeans
inertia=[]
k_values=range(2,15)
for i in k_values:
  km=KMeans(n_clusters=i,random_state=42)
  km.fit(pc_array1)
  css=km.inertia_
  inertia.append(css)
plt.plot(k_values,inertia,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python

# Identify the optimal K using the "elbow" method
diff = np.diff(inertia)
k_optimal = k_values[np.argmin(diff) + 1]

print(f"The optimal number of clusters (K) is: {k_optimal}")
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

silhouette_scores = []
K = range(2, 16)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(pc_array1)
    score = silhouette_score(pc_array1, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"K={k}, Silhouette Score={score:.4f}")

# Find the K with the highest silhouette score
best_k = K[silhouette_scores.index(max(silhouette_scores))]
print(f"\nOptimal number of clusters based on silhouette score: {best_k}")
```

```python

#Creating k-means clustering model
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
y_kmeans = kmeans.fit_predict(pc_array1)
```

```python
# Assuming k_optimal is 3.
plt.figure(figsize=(8,6))
for i in range(k_optimal):
    m= (y_kmeans == i)
    plt.scatter(pc_array1.iloc[m, 0], pc_array1.iloc[m, 1], s=100, label=f'Cluster {i+1}')

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='yellow', marker='*', label='Centroids')

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='yellow', marker='*', label='Centroids')

plt.title('Clusters of Customers (K-means Clustering)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python

#calculating silhouette_score
from sklearn.metrics import silhouette_score
score = silhouette_score(pc_array1, y_kmeans)
print(f"Silhouette Score: {score:.2f}")
```

```python
#DBSCAN
#creating DBSCAN model
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.3,min_samples=3)
y_db=dbscan.fit_predict(pc_array1)
```

```python

unique_labels = set(y_db)
colors = plt.cm.get_cmap('tab10', len(unique_labels))
```

```python
#plot clusters
plt.figure(figsize=(8, 6))
for label in unique_labels:
    if label == -1:
        color = 'gray'
        label_name = 'Noise'
    else:
        color = colors(label)
        label_name = f'Cluster {label + 1}'
    m3 = (y_db == label)
    plt.scatter(
        pc_array1.loc[m3, pc_array1.columns[0]],
        pc_array1.loc[m3, pc_array1.columns[1]],
        s=100, c=color, label=label_name
    )
plt.title('Clusters of Customers (DBSCAN)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python

#calculating silhouette_score
score = silhouette_score(pc_array1, y_db)
print(f"Silhouette Score (DBSCAN): {score:.4f}")
```
