```python
#impute Numerical column with median value
fill_cols=['LoanAmount','Loan_Amount_Term','Credit_History']
for col in fill_cols:
    median_value=train_loan[col].median()
    train_loan[col].fillna(median_value,inplace=True)
```

```python
#impute Categorical column with mode value
fill_cols=['Gender','Married','Dependents','Self_Employed']
for col in fill_cols:
    mode_value=train_loan[col].mode()[0]
    train_loan[col].fillna(mode_value,inplace=True)
```

```python
#Apply IQR method
cols=['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
for c in cols:
    Q1=train_loan[c].quantile(0.25)
    Q3=train_loan[c].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    train_loan[c]=np.where(train_loan[c]<lower_bound,lower_bound,train_loan[c])
    train_loan[c]=np.where(train_loan[c]>upper_bound,upper_bound,train_loan[c])
```

Encoding

```python
#Encoding

train_loan['Loan_Status']=train_loan['Loan_Status'].map({'Y':1,'N':0})
#train_loan['Property_Area']=train_loan['Property_Area'].map(size_mapping)
train_loan.head()
```

```python
train_loan['Dependents']=train_loan['Dependents'].replace('3+',3,inplace=True)
train_loan.head()
```

```python
#Encoding
#1.Label Encoding
#2.One-hot Encoding
#import Label Encoder
from sklearn.preprocessing import LabelEncoder
binary_col=['Gender','Married','Education','Self_Employed']

# Initialize the LabelEncoder
label_encoder = {}
for col in binary_col:
    le = LabelEncoder()
    train_loan[col] = le.fit_transform(train_loan[col])
    label_encoder[col] = le
train_loan.head(10)
```

```python
print(train_loan['Property_Area'].value_counts())
print(train_loan['Property_Area'].isnull().sum())
```

```python
print(train_loan['Dependents'].value_counts())
print(train_loan['Dependents'].isnull().sum())
```

```python
#One hot Encoding
train_loan=pd.get_dummies(train_loan,columns=['Dependents','Property_Area'])
train_loan.head()
```

```python
#Drop Column Loan_ID
train_loan.drop('Loan_ID',axis=1,inplace=True)
train_loan.head()
```

```python
train_loan.columns
```

```python
#Convert true or false column to 1 or 0
train_loan=train_loan.astype('int64')
train_loan.head(3)
```

Split the data into training & Testing set

```python
# Importing libraries for Train-test split
from sklearn.model_selection import train_test_split
# Split the data into features (x) and target variable (y)
x = train_loan.drop('Loan_Status', axis=1)
y = train_loan['Loan_Status']
```

```python
print(x)
```

```python
print(y)
```

```python
#Split data into training and test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train=x_train_sc
x_test=x_test_sc
```

**Ensemble Learning Technique**

```python
#import libraries for ensemble learning techniques
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import statistics as st
```

```python
pre1=model1.predict(x_test)
pre2=model2.predict(x_test)
pre3=model3.predict(x_test)
```

HyperParameter Tuning

```python

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0]
}
modelog=GradientBoostingClassifier(learning_rate=0.01,n_estimators=5,random_state=42)


grid_search = GridSearchCV(
    estimator=modelog,
    param_grid=param_grid,
    cv=5,
    verbose=2,
    n_jobs=-1)
grid_search.fit(x_train, y_train)

print("Best parameters found by GridSearchCV:", grid_search.best_params_)
modelog_best = grid_search.best_estimator_
modelog_pred = modelog_best.predict(x_test)
```
