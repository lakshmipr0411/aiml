```python
# Columns to apply the IQR method
numeric_col = ['no_of_trainings','age','previous_year_rating','length_of_service','awards_won?']

# Applying the IQR method to each specified column
for col in numeric_col:
    Q1 = train_data[col].quantile(0.25)
    Q3 = train_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    train_data[col] = np.where(train_data[col] < lower_bound, lower_bound, train_data[col])
    train_data[col] = np.where(train_data[col] > upper_bound, upper_bound, train_data[col])
```
