Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5,5))
plt.hist(diabetes['Age'],color='blue',edgecolor='y')
plt.xlabel('Age of Patients')
plt.ylabel('Frequency')
plt.show()
```

```python
# change Outcome as category type
diabetes['Outcome']= diabetes['Outcome'].astype('category')
diabetes.info()
```

Boxplot

```python
#age distribution by Outcome (x= 'Outcome' y='Age)
plt.figure(figsize=(5,3))
sns.boxplot(x='Outcome',y='Age',data=diabetes,palette='coolwarm')
plt.xlabel('Outcome')
plt.ylabel('Age')
plt.title('Age distribution by Outcome')
plt.show()
```

Scatterplot

```python
#age distribution by Insulin (x= 'Insulin' y='Age)
plt.figure(figsize=(5,3))
sns.scatterplot(x='Insulin',y='Age',data=diabetes,color='red')
plt.xlabel('Insulin')
plt.ylabel('Age')
plt.title('Age distribution by Insulin')
plt.show()
```
