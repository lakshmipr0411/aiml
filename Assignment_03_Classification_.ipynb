##Data Preprocessing

```python
#Check the null values
diabetes.isnull().sum()
```

```python
#To change the Zero's to nan value because Some column having 0 in diabetes dataset
diabetes[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]=diabetes[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].replace(0,np.nan)
```

```python
diabetes.isnull().sum()
```

```python
#Filling the missing values using median imputation because the columns are Numerical
cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
for c in cols:
  median_value=diabetes[c].median()
  diabetes[c].fillna(median_value,inplace=True)
```

```python
#After imputation
diabetes.isnull().sum()
```

```python
# Columns to apply the IQR method
numeric_col = ['Pregnancies','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']

# Applying the IQR method to each specified column
for col in numeric_col:
    Q1 = diabetes[col].quantile(0.25)
    Q3 = diabetes[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    diabetes[col] = np.where(diabetes[col] < lower_bound, lower_bound, diabetes[col])
    diabetes[col] = np.where(diabetes[col] > upper_bound, upper_bound, diabetes[col])
```

```python
#list the columns for box plot After Outlier Treatment
box_plot_columns = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
# Create boxplot
plt.figure(figsize=(8,5))
diabetes[box_plot_columns].boxplot()
plt.title('Boxplot of Multiple Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=90)
plt.show()
```

##Model Development

```python
# Importing libraries for Train-test split,cross_value_score

from sklearn.model_selection import train_test_split, cross_val_score
# Split the data into features (x) and target variable (y)
x = diabetes.drop('Outcome', axis=1)
y =diabetes['Outcome']
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

Different Types of Models

**Logistic regression**



```python
#Logistic Regression Classifier
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression(random_state=42,max_iter=50000)
lr.fit(x_train,y_train)
#y Prediction
y_pred=lr.predict(x_test)
y_pred
```

```python
#Importing confusion matrix libraries
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
# Importing Cross validation Libraries

from sklearn.model_selection import cross_val_score

print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy Score:",accuracy_score(y_pred,y_test))
print('\n Classification Report: \n', classification_report(y_pred, y_test))
cv_scores = cross_val_score(lr, x_train, y_train, cv=5)
print('Cross Validation Scores:', cross_val_score(lr,x_train,y_train,cv=5))
```

```python

#after scaling
lr.fit(x_train_sc,y_train)
y_pred=lr.predict(x_test_sc)
```

```python
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy Score:",accuracy_score(y_pred,y_test))
print('\n Classification Report: \n', classification_report(y_pred, y_test))
cv_scores = cross_val_score(lr, x_train, y_train, cv=5)
print('Cross Validation Scores:', cross_val_score(lr,x_train,y_train,cv=5))
```

**K-Nearest Neighbors**
